# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

## Answer:
**Matplotlib** is a widely used plotting library for the Python programming language. It provides a variety of functions for creating static, animated, and interactive visualizations in Python. Matplotlib is particularly popular in scientific computing and data analysis for visualizing data, performing exploratory data analysis (EDA), and creating publication-quality graphics.

### It is used for:
- **Data visualization**: Plotting graphs to help in data analysis and interpretation.
- **Customization**: Providing control over every aspect of the plot, from line styles to axis labels.
- **Flexibility**: Creating a wide variety of plot types (such as bar charts, histograms, scatter plots, etc.).

### Five plots that can be plotted using the Pyplot module of Matplotlib:
1. **Line Plot**: A basic plot to visualize data points connected by straight lines.
2. **Bar Plot**: Used for categorical data to display values as bars.
3. **Histogram**: Used to visualize the distribution of numerical data.
4. **Scatter Plot**: Displays points based on two variables, typically used to show relationships between variables.
5. **Pie Chart**: Used to represent proportions of a whole, where each slice represents a category.


# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.  
import numpy as np  
np.random.seed(3)  
x = 3 + np.random.normal(0, 2, 50)  
y = 3 + np.random.normal(0, 2, len(x))  
Note: Also add title, xlabel, and ylabel to the plot.

## Answer:
A scatter plot is a type of plot that uses dots to represent the values obtained from two variables. It is mainly used to visualize relationships between two numerical variables, where one variable is plotted along the x-axis and the other along the y-axis.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plotting the scatter plot
plt.scatter(x, y)

# Adding title and labels
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")

# Show the plot
plt.show()
```

# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.  

### Answer:
The **subplot()** function in Matplotlib is used to create multiple plots in a single figure. It allows us to define the number of rows and columns of subplots in the figure, and also to select the specific position of each plot. This is particularly useful for comparing multiple graphs in one view.

### Code to draw four line plots using the subplot() function:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1 data
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2 data
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3 data
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4 data
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create 2x2 subplot
plt.subplot(2, 2, 1)  # First subplot in a 2x2 grid
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)  # Second subplot in a 2x2 grid
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)  # Third subplot in a 2x2 grid
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)  # Fourth subplot in a 2x2 grid
plt.plot(x, y4)
plt.title("Line 4")

# Display the plots
plt.tight_layout()
plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

### Answer:
A **bar plot** (or bar chart) is a type of data visualization that uses rectangular bars to represent the values of different categories. The length of each bar is proportional to the value it represents. Bar plots are used to compare different categories or show the frequency distribution of a set of data.

Bar plots are particularly useful for:
- Comparing quantities across different categories.
- Visualizing differences in values in a clear and concise way.

### Code to plot a bar plot and a horizontal bar plot:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit)
plt.title("Company Profits (Bar Plot)")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title("Company Profits (Horizontal Bar Plot)")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

### Answer:
A **box plot** (also known as a **box-and-whisker plot**) is a statistical graphic that displays the distribution of a dataset. It shows the dataset's minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum, as well as any potential outliers. The box plot is used to provide a summary of the distribution, allowing us to easily see the central tendency, variability, and presence of outliers in the data.

Box plots are particularly useful for:
- Identifying the spread and skewness of data.
- Comparing distributions between multiple datasets.
- Detecting outliers in the data.

### Code to plot a box plot:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)  # Mean = 100, Std = 10, n = 200
box2 = np.random.normal(90, 20, 200)   # Mean = 90, Std = 20, n = 200

# Creating box plot
plt.boxplot([box1, box2], labels=["Box 1", "Box 2"])
plt.title("Comparison of Two Box Plots")
plt.ylabel("Values")
plt.show()
```
