# Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

### Answer:

1. **Line Plot**:
   - **Use**: A line plot is used to visualize the relationship between two continuous variables. It is ideal for showing trends over time.
   - **Example Use Case**: Visualizing stock prices over a period of time.

2. **Bar Plot**:
   - **Use**: A bar plot is used to compare different categories or groups. It displays data with rectangular bars where the length of each bar is proportional to the value it represents.
   - **Example Use Case**: Comparing the sales of different products in a given year.

3. **Box Plot**:
   - **Use**: A box plot is used to show the distribution of a dataset. It highlights the median, quartiles, and outliers, which helps to understand the spread and skewness of data.
   - **Example Use Case**: Visualizing the distribution of test scores of a class.

4. **Heatmap**:
   - **Use**: A heatmap is used to visualize data in a matrix form, where individual values are represented by colors. It is often used to plot correlation matrices or categorical data.
   - **Example Use Case**: Visualizing correlation between multiple variables in a dataset.

5. **Pair Plot**:
   - **Use**: A pair plot is used to visualize relationships between pairs of variables in a dataset. It creates a grid of scatterplots for each pair of features.
   - **Example Use Case**: Exploring relationships between multiple features in the Iris dataset (such as petal length, petal width, etc.).

### These plots help in quickly understanding patterns, distributions, and relationships in datasets, making Seaborn a powerful tool for exploratory data analysis.


# Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

### Answer:

To load the "fmri" dataset using Seaborn and create a line plot with the specified x and y values, you can follow these steps:

1. First, we load the dataset using the `load_dataset` function.
2. Then, we use the `lineplot` function from Seaborn to plot the data, with "timepoint" on the x-axis and "signal" on the y-axis, grouped by different "event" and "region".

### Code

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create a line plot
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Display the plot
plt.show()
```

# Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

### Answer:

To load the "titanic" dataset and create box plots for the columns 'age' and 'fare' based on the 'pclass', you can follow these steps:

1. Load the dataset using Seaborn's `load_dataset` function.
2. Create two box plots using the `boxplot` function:
   - The first box plot will show 'age' against 'pclass'.
   - The second box plot will show 'fare' against 'pclass'.

Here’s the code for creating the box plots:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Set up the matplotlib figure
plt.figure(figsize=(12, 6))

# Create the first box plot for 'age' vs 'pclass'
plt.subplot(1, 2, 1)  # Create subplot (1 row, 2 columns, first plot)
sns.boxplot(data=titanic, x="pclass", y="age")
plt.title('Boxplot of Age vs Pclass')

# Create the second box plot for 'fare' vs 'pclass'
plt.subplot(1, 2, 2)  # Create subplot (1 row, 2 columns, second plot)
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title('Boxplot of Fare vs Pclass')

# Display the plots
plt.tight_layout()
plt.show()
```

# Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

### Answer:

To create a histogram for the 'price' column from the "diamonds" dataset and differentiate the data by the 'cut' column using the `hue` parameter, follow these steps:

1. Load the "diamonds" dataset using Seaborn’s `load_dataset` function.
2. Create the histogram using the `histplot` function, setting 'price' as the variable to plot and using the 'cut' column as the hue to differentiate the categories.

Here’s the code to generate the histogram:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create the histogram
sns.histplot(data=diamonds, x="price", hue="cut", kde=True)

# Set the title and labels
plt.title('Histogram of Diamond Prices with Cut as Hue')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

# Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

### Answer:

To create a pair plot from the "iris" dataset and differentiate the data by the 'species' column using the `hue` parameter, follow these steps:

1. Load the "iris" dataset using Seaborn’s `load_dataset` function.
2. Create the pair plot using the `pairplot` function, setting 'species' as the hue to color-code the data by species.

Here’s the code to generate the pair plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the iris dataset
iris = sns.load_dataset("iris")

# Create the pair plot
sns.pairplot(iris, hue="species")

# Set the title
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)

# Display the plot
plt.show()
```

# Q6: Use the "flights" dataset from seaborn to plot a heatmap.

### Answer:

To create a heatmap using the "flights" dataset, follow these steps:

1. Load the "flights" dataset using Seaborn’s `load_dataset` function.
2. Pivot the dataset to get it into a suitable format for the heatmap.
3. Use Seaborn’s `heatmap` function to create the heatmap.

Here’s the code to generate the heatmap:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix for the heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Create the heatmap
sns.heatmap(flights_pivot, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=0.5)

# Set the title
plt.title('Heatmap of Flight Passengers')

# Display the plot
plt.show()
```
