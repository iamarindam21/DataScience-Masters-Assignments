#Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

### 1. **t-test**:
   - Used when the **sample size is small** (\( n < 30 \)) and/or the **population standard deviation (\( \sigma \)) is unknown**.
   - Based on the **Student's t-distribution**, which has heavier tails than the normal distribution.
   - As sample size increases, the t-distribution approaches the normal distribution.

   **Example Scenario**:
   - A researcher wants to compare the **mean test scores** of a sample of **20 students** to a known population mean. Since the sample size is small and the population standard deviation is unknown, a **t-test** is appropriate.

### 2. **z-test**:
   - Used when the **sample size is large** (\( n \geq 30 \)) and the **population standard deviation (\( \sigma \)) is known**.
   - Based on the **standard normal distribution (Z-distribution)**.
   - Assumes that the data follows a normal distribution, or the sample size is large enough for the Central Limit Theorem to apply.

   **Example Scenario**:
   - A factory produces light bulbs, and the manufacturer wants to test whether the **average lifespan** of the bulbs is 1000 hours. They take a sample of **100 bulbs**, and since the **population standard deviation is known**, a **z-test** is appropriate.


# Q2: Differentiate between one-tailed and two-tailed tests.
### 1. **One-Tailed Test**:
   - Tests whether the sample mean is **significantly greater than or less than** the population mean.
   - The critical region (rejection area) is located in **only one tail** of the distribution.
   - Used when we have a **specific directional hypothesis** (e.g., "greater than" or "less than").

   #### **Example Scenario**:
   - A pharmaceutical company wants to test whether a new drug **increases** blood pressure. The hypothesis is one-sided:  
     - **Null Hypothesis** (\( H_0 \)): The drug has no effect or lowers blood pressure.  
     - **Alternative Hypothesis** (\( H_A \)): The drug **increases** blood pressure.  
   - A **one-tailed test** is used because we are only interested in an **increase**.

### 2. **Two-Tailed Test**:
   - Tests whether the sample mean is **significantly different** (either greater or smaller) from the population mean.
   - The critical region (rejection area) is located in **both tails** of the distribution.
   - Used when we have a **non-directional hypothesis** (e.g., "different from").

 ####  **Example Scenario**:
   - A manufacturer wants to test whether a machine produces bolts with an **average length of 5 cm**. They are concerned about **both shorter and longer bolts**:
     - **Null Hypothesis** (\( H_0 \)): The average bolt length is 5 cm.  
     - **Alternative Hypothesis** (\( H_A \)): The average bolt length is **not** 5 cm.  
   - A **two-tailed test** is used because we are interested in **both increases and decreases**.


# Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

In hypothesis testing, errors occur when we make incorrect decisions about the null hypothesis (\( H_0 \)). The two main types of errors are **Type 1 Error** and **Type 2 Error**.

### 1. **Type 1 Error (False Positive)**
   - Occurs when we **reject** the null hypothesis (\( H_0 \)) when it is actually **true**.
   - The probability of making a Type 1 Error is denoted by **\( \alpha \) (significance level)**.
   - A **lower \( \alpha \)** reduces the risk of a Type 1 Error but increases the risk of a Type 2 Error.

   **Example Scenario**:
   - A court trial follows the assumption that a defendant is **innocent until proven guilty**.
   - **Type 1 Error:** The court **wrongly convicts an innocent person**.
   - In this case, the null hypothesis (\( H_0 \)) is "The defendant is innocent," but it is wrongly rejected.

### 2. **Type 2 Error (False Negative)**
   - Occurs when we **fail to reject** the null hypothesis (\( H_0 \)) when it is actually **false**.
   - The probability of making a Type 2 Error is denoted by **\( \beta \)**.
   - A **lower Type 2 Error risk** increases the test's **power (\( 1 - \beta \))**, making it more sensitive to detecting real effects.

   **Example Scenario**:
   - In medical testing, a new diagnostic test is developed to detect a disease.
   - **Type 2 Error:** The test **fails to detect the disease** in a sick patient.
   - Here, the null hypothesis (\( H_0 \)) is "The patient does not have the disease," and it is wrongly accepted.


# Q4: Explain Bayes's theorem with an example.

## Bayes's Theorem

Bayes's Theorem describes the probability of an event based on prior knowledge of related conditions. It is mathematically expressed as:

\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]

Where:
- \( P(A|B) \) = Probability of event \( A \) given \( B \) (posterior probability).
- \( P(B|A) \) = Probability of event \( B \) given \( A \) (likelihood).
- \( P(A) \) = Prior probability of event \( A \).
- \( P(B) \) = Total probability of event \( B \).

### **Example Scenario: Medical Testing**
Suppose a disease affects **1%** of a population (\( P(A) = 0.01 \)). A test for the disease is **90% accurate for positive cases** (\( P(B|A) = 0.9 \)) and **95% accurate for negative cases** (\( P(B|\neg A) = 0.05 \)).

We want to find the probability that a person actually has the disease given that they tested positive (\( P(A|B) \)).

Using Bayes's Theorem:

\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]

First, calculate \( P(B) \) (the total probability of a positive test):

\[
P(B) = P(B|A) P(A) + P(B|\neg A) P(\neg A)
\]

\[
P(B) = (0.9 \times 0.01) + (0.05 \times 0.99) = 0.009 + 0.0495 = 0.0585
\]

Now, applying Bayes’s formula:

\[
P(A|B) = \frac{0.9 \times 0.01}{0.0585} = \frac{0.009}{0.0585} \approx 0.154
\]

### **Interpretation**
Even with a **positive test result**, the probability of actually having the disease is **only 15.4%**, not 90%. This happens because the disease is rare, and false positives can significantly impact the result.

### **Use Cases of Bayes's Theorem**
- **Medical diagnostics** (as shown in the example).
- **Spam filtering** (determining if an email is spam based on words used).
- **Machine learning** (e.g., Naïve Bayes classifier).
- **Risk assessment** (fraud detection, security analysis).



# Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.
# Confidence Interval (CI)

A **confidence interval (CI)** is a range of values that is likely to contain the true population parameter (e.g., mean or proportion) with a certain level of confidence. It is used to estimate population parameters based on sample data.

### **Formula for Confidence Interval**
For a population mean (\( \mu \)) with known standard deviation (\( \sigma \)), the confidence interval is given by:

\[
CI = \bar{x} \pm Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
\]

Where:
- \( \bar{x} \) = Sample mean
- \( Z_{\alpha/2} \) = Critical value from the Z-table (for large samples) or t-table (for small samples)
- \( \sigma \) = Population standard deviation (or sample standard deviation if \( n \) is small)
- \( n \) = Sample size

### **Example Calculation**
A researcher collects a sample of **50 students** to estimate their **average height**. The **sample mean height** is **170 cm**, and the **standard deviation** is **10 cm**. We calculate a **95% confidence interval** for the population mean.

#### **Step 1: Identify Values**
- \( \bar{x} = 170 \)
- \( \sigma = 10 \)
- \( n = 50 \)
- **For a 95% confidence level**, the **Z-value** from the Z-table is **1.96**.

#### **Step 2: Compute the Margin of Error**
\[
ME = 1.96 \times \frac{10}{\sqrt{50}}
\]

\[
ME = 1.96 \times \frac{10}{7.07} \approx 1.96 \times 1.41 \approx 2.76
\]

#### **Step 3: Compute the Confidence Interval**
\[
CI = 170 \pm 2.76
\]

\[
CI = (167.24, 172.76)
\]

### **Interpretation**
We are **95% confident** that the **true average height** of all students is between **167.24 cm and 172.76 cm**.


# Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the event's probability and new evidence. Provide a sample problem and solution.

## Bayes' Theorem: Probability Update with New Evidence

Bayes' Theorem allows us to update the probability of an event occurring based on prior knowledge and new evidence.

### **Bayes' Theorem Formula**
\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]
Where:
- \( P(A|B) \) = Probability of event \( A \) given that \( B \) has occurred (posterior probability).
- \( P(B|A) \) = Probability of event \( B \) given that \( A \) has occurred (likelihood).
- \( P(A) \) = Prior probability of event \( A \).
- \( P(B) \) = Total probability of event \( B \) occurring.

---

## **Example Problem: Disease Diagnosis**
A hospital is testing for a rare disease that affects **2%** of the population (\( P(A) = 0.02 \)). The test is:
- **95% accurate** for detecting the disease when a person has it (\( P(B|A) = 0.95 \)).
- **10% false positive rate**, meaning 10% of healthy people incorrectly test positive (\( P(B|\neg A) = 0.10 \)).

If a person tests **positive**, what is the probability they **actually have** the disease (\( P(A|B) \))?

---

### **Step 1: Compute \( P(B) \) (Total Probability of Testing Positive)**

Using the **Law of Total Probability**:

\[
P(B) = P(B|A) P(A) + P(B|\neg A) P(\neg A)
\]

\[
P(B) = (0.95 \times 0.02) + (0.10 \times 0.98)
\]

\[
P(B) = 0.019 + 0.098 = 0.117
\]

---

### **Step 2: Apply Bayes’ Theorem**

\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]

\[
P(A|B) = \frac{0.95 \times 0.02}{0.117}
\]

\[
P(A|B) = \frac{0.019}{0.117} \approx 0.162
\]

---

### **Final Answer & Interpretation**
If a person tests **positive**, the probability that they **actually have** the disease is **16.2%**, not 95%! This is because false positives significantly affect the result when the disease is rare.

# Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

## Margin of Error in a Confidence Interval

The **margin of error (ME)** represents the amount of uncertainty in an estimate from a sample. It is the range added and subtracted from the sample statistic (e.g., mean) to create a confidence interval.

### **Formula for Margin of Error**
\[
ME = Z_{\alpha/2} \times \frac{\sigma}{\sqrt{n}}
\]
Where:
- \( Z_{\alpha/2} \) = Critical value from the Z-table (for a given confidence level)
- \( \sigma \) = Population standard deviation (or sample standard deviation if unknown)
- \( n \) = Sample size

### **Effect of Sample Size on Margin of Error**
- **Larger sample size (\( n \)) → Smaller margin of error**
- **Smaller sample size (\( n \)) → Larger margin of error**

Since **\( ME \propto \frac{1}{\sqrt{n}} \)**, increasing the sample size reduces the uncertainty, making the confidence interval narrower.

### **Example Scenario**
A survey is conducted to estimate the **average height** of adults in a city. Two scenarios:

1. **Small Sample (\( n = 25 \))**
   - Sample mean: 170 cm
   - Standard deviation: 10 cm
   - 95% confidence level (\( Z = 1.96 \))

   \[
   ME = 1.96 \times \frac{10}{\sqrt{25}} = 1.96 \times 2 = 3.92
   \]

   **Confidence Interval:** \( 170 \pm 3.92 \) → \( (166.08, 173.92) \)

2. **Larger Sample (\( n = 100 \))**
   - Everything remains the same except \( n \).

   \[
   ME = 1.96 \times \frac{10}{\sqrt{100}} = 1.96 \times 1 = 1.96
   \]

   **Confidence Interval:** \( 170 \pm 1.96 \) → \( (168.04, 171.96) \)

### **Conclusion**
- **With a larger sample (\( n = 100 \)), the margin of error decreases from 3.92 to 1.96**, making the estimate **more precise**.
- **Larger samples reduce uncertainty and improve confidence interval accuracy**.

### **Real-World Example**
- In **political polling**, a survey of **1000 voters** gives a **smaller margin of error** than a survey of **100 voters**, leading to **more reliable election predictions**.


# Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

## Z-Score Calculation

The **Z-score** measures how many standard deviations a data point is from the **mean**. It is calculated using the formula:

\[
Z = \frac{X - \mu}{\sigma}
\]

Where:
- \( X \) = Data point (75)
- \( \mu \) = Population mean (70)
- \( \sigma \) = Population standard deviation (5)

### **Step 1: Compute the Z-score**
\[
Z = \frac{75 - 70}{5} = \frac{5}{5} = 1.0
\]

### **Step 2: Interpretation of the Z-score**
- A **Z-score of 1.0** means that the data point (75) is **1 standard deviation above the mean**.
- In a **normal distribution**, about **68% of the data** falls within **one standard deviation** from the mean (\( -1 \leq Z \leq 1 \)).
- This suggests that **75 is not an extreme value** and is fairly typical within the dataset.

### **Real-World Example**
- If this were a **test score**, where the average score is **70** with a **standard deviation of 5**, a score of **75** would be slightly above average, but not exceptionally high.


# Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an averageof 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

## Hypothesis Test for Weight Loss Drug Effectiveness (t-test)

We will perform a **one-sample t-test** to determine if the drug is significantly effective.

---

### **Step 1: Define Hypotheses**
- **Null Hypothesis (\( H_0 \))**: The drug is **not effective** (i.e., the true mean weight loss is **0 pounds**).
  \[
  H_0: \mu = 0
  \]
- **Alternative Hypothesis (\( H_A \))**: The drug is **effective** (i.e., the true mean weight loss is **greater than 0 pounds**).
  \[
  H_A: \mu > 0
  \]
This is a **one-tailed test** since we are only interested in whether the weight loss is significantly **greater than zero**.

---

### **Step 2: Given Data**
- **Sample Mean (\( \bar{x} \))** = 6 pounds
- **Standard Deviation (\( s \))** = 2.5 pounds
- **Sample Size (\( n \))** = 50
- **Significance Level (\( \alpha \))** = 0.05 (95% confidence level)
- **Degrees of Freedom (\( df \))** = \( n - 1 = 50 - 1 = 49 \)

---

### **Step 3: Compute the Test Statistic (t-score)**
\[
t = \frac{\bar{x} - \mu}{s / \sqrt{n}}
\]

\[
t = \frac{6 - 0}{2.5 / \sqrt{50}}
\]

\[
t = \frac{6}{2.5 / 7.07} = \frac{6}{0.353} \approx 17.00
\]

---

### **Step 4: Determine the Critical Value**
Using a **t-table**, for **\( df = 49 \)** and **\( \alpha = 0.05 \) (one-tailed test)**:
- **Critical t-value (\( t_{critical} \))** ≈ **1.676**

---

### **Step 5: Compare t-score with Critical Value**
- **Computed t-score** = **17.00**
- **Critical t-value** = **1.676**
- Since **\( t > t_{critical} \) (17.00 > 1.676)**, we **reject the null hypothesis**.

---

### **Step 6: Conclusion**
Since the computed t-score **far exceeds** the critical value, we conclude that:

 **There is strong statistical evidence that the weight loss drug is effective.**  
At a **95% confidence level**, the drug **significantly reduces weight**.

---

### **Real-World Implication**
- The drug has a meaningful effect on weight loss.
- Future studies could compare it to a **placebo group** to further confirm its effectiveness.


# Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95% confidence interval for the true proportion of people who are satisfied with their job.

## 95% Confidence Interval for Proportion

Given:
- **Sample size (\( n \))** = 500
- **Sample proportion (\( \hat{p} \))** = 65% = 0.65
- **Confidence level** = 95%

### **Step 1: Compute the Z-value**
For a 95% confidence level, the **Z-value** from the Z-table is **1.96**.

### **Step 2: Calculate the Standard Error (SE) for Proportion**
The formula for the standard error of a proportion is:

\[
SE = \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}
\]

Substitute the given values:

\[
SE = \sqrt{\frac{0.65(1 - 0.65)}{500}} = \sqrt{\frac{0.65 \times 0.35}{500}} = \sqrt{\frac{0.2275}{500}} = \sqrt{0.000455} \approx 0.0213
\]

### **Step 3: Compute the Margin of Error (ME)**
The margin of error is calculated as:

\[
ME = Z_{\alpha/2} \times SE
\]

\[
ME = 1.96 \times 0.0213 \approx 0.0417
\]

### **Step 4: Compute the Confidence Interval**
The confidence interval for the proportion is:

\[
CI = \hat{p} \pm ME
\]

\[
CI = 0.65 \pm 0.0417
\]

\[
CI = (0.6083, 0.6917)
\]

### **Interpretation**
- We are **95% confident** that the true proportion of people who are satisfied with their job lies between **60.83% and 69.17%**.
- This means that, based on the survey, we expect that between **60.83%** and **69.17%** of the entire population of people would report being satisfied with their current job.

### **Real-World Implication**
- The survey results suggest a **strong majority** of people are satisfied with their job, but the confidence interval allows for uncertainty, reflecting the sample's margin of error.


# Q12. A researcher is testing the effectiveness of two different teaching methods on student performance.Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. Conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.

## Hypothesis Test for Comparing Two Teaching Methods (Two-Sample t-test)

We will perform a **two-sample t-test** to determine if there is a significant difference in student performance between two teaching methods.

---

### **Step 1: Define Hypotheses**

- **Null Hypothesis (\( H_0 \))**: There is no difference in student performance between the two teaching methods. (\( \mu_A = \mu_B \))
  
  \[
  H_0: \mu_A = \mu_B
  \]

- **Alternative Hypothesis (\( H_A \))**: There is a significant difference in student performance between the two teaching methods. (\( \mu_A \neq \mu_B \))

  \[
  H_A: \mu_A \neq \mu_B
  \]

This is a **two-tailed test** because we are testing for any difference (greater or smaller) between the means.

---

### **Step 2: Given Data**
- **Sample A**: Mean (\( \bar{X}_A \)) = 85, Standard Deviation (\( s_A \)) = 6
- **Sample B**: Mean (\( \bar{X}_B \)) = 82, Standard Deviation (\( s_B \)) = 5
- **Significance Level (\( \alpha \))** = 0.01
- **Sample sizes**: Assume **n₁ = n₂ = 30** (for simplicity)

---

### **Step 3: Calculate the Test Statistic (t-value)**

The formula for the two-sample t-test is:

\[
t = \frac{(\bar{X}_A - \bar{X}_B)}{\sqrt{\frac{s_A^2}{n_A} + \frac{s_B^2}{n_B}}}
\]

Substitute the given values:

\[
t = \frac{(85 - 82)}{\sqrt{\frac{6^2}{30} + \frac{5^2}{30}}} = \frac{3}{\sqrt{\frac{36}{30} + \frac{25}{30}}} = \frac{3}{\sqrt{2.2 + 1.67}} = \frac{3}{\sqrt{3.87}} = \frac{3}{1.97} \approx 1.52
\]

---

### **Step 4: Degrees of Freedom and Critical Value**

For a **two-tailed test** at a significance level of **0.01**, we need to find the **critical t-value**.

- **Degrees of Freedom (\( df \))**: For equal sample sizes, we can use the formula:
  
  \[
  df = n_A + n_B - 2 = 30 + 30 - 2 = 58
  \]

- Using a **t-distribution table** or **t-distribution calculator**, the **critical t-value** for \( df = 58 \) and \( \alpha = 0.01 \) (two-tailed) is approximately **2.660**.

---

### **Step 5: Compare t-statistic with Critical Value**

- **Computed t-value** = **1.52**
- **Critical t-value** = **±2.660**

Since **1.52 < 2.660**, we **fail to reject the null hypothesis**.

---

### **Step 6: Conclusion**

There is **insufficient evidence** to suggest that there is a **significant difference** in student performance between the two teaching methods at the **0.01 significance level**.

### **Interpretation**
- Although there is a **difference in means (85 vs. 82)**, the difference is not statistically significant given the **sample size** and **variability** in scores.

### **Real-World Implication**
- The two teaching methods appear to result in similar student performance based on this analysis. Future studies might consider **larger sample sizes** or **different measures of student performance** to draw a more definitive conclusion.


# Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 90% confidence interval for the true population mean.

## 90% Confidence Interval for Population Mean

Given:
- **Population Mean (\( \mu \))** = 60
- **Population Standard Deviation (\( \sigma \))** = 8
- **Sample Mean (\( \bar{x} \))** = 65
- **Sample Size (\( n \))** = 50
- **Confidence Level** = 90%

### **Step 1: Determine the Z-value**
For a 90% confidence level, the **Z-value** (critical value) is **1.645** (from Z-table for two-tailed test).

### **Step 2: Compute the Standard Error (SE)**
The formula for standard error of the mean is:

\[
SE = \frac{\sigma}{\sqrt{n}}
\]

Substitute the given values:

\[
SE = \frac{8}{\sqrt{50}} = \frac{8}{7.071} \approx 1.131
\]

### **Step 3: Compute the Margin of Error (ME)**
The margin of error is calculated as:

\[
ME = Z_{\alpha/2} \times SE
\]

\[
ME = 1.645 \times 1.131 \approx 1.858
\]

### **Step 4: Compute the Confidence Interval**
The confidence interval for the true population mean is:

\[
CI = \bar{x} \pm ME
\]

\[
CI = 65 \pm 1.858
\]

\[
CI = (65 - 1.858, 65 + 1.858) = (63.142, 66.858)
\]

### **Interpretation**
- We are **90% confident** that the true population mean lies between **63.142 and 66.858**.
- This means that based on the sample, we expect the true population mean to fall within this range with 90% certainty.

### **Real-World Implication**
- If we were to repeat this sampling process many times, we would expect the true population mean to lie within this interval in **90% of the samples**.


# Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

## Hypothesis Test for the Effect of Caffeine on Reaction Time (t-test)

We will perform a **one-sample t-test** to determine if caffeine has a significant effect on reaction time. We will compare the sample mean reaction time to the **hypothesized population mean** (e.g., no caffeine effect).

---

### **Step 1: Define Hypotheses**

- **Null Hypothesis (\( H_0 \))**: Caffeine has **no effect** on reaction time, so the mean reaction time is equal to a hypothesized value (e.g., 0.30 seconds, assuming no caffeine effect).
  
  \[
  H_0: \mu = 0.30 \text{ seconds}
  \]

- **Alternative Hypothesis (\( H_A \))**: Caffeine has a **significant effect** on reaction time, so the mean reaction time is **not equal to 0.30 seconds**.
  
  \[
  H_A: \mu \neq 0.30 \text{ seconds}
  \]

This is a **two-tailed test** because we are testing for any difference in reaction time (either an increase or decrease).

---

### **Step 2: Given Data**
- **Sample Mean (\( \bar{x} \))** = 0.25 seconds
- **Standard Deviation (\( s \))** = 0.05 seconds
- **Sample Size (\( n \))** = 30
- **Significance Level (\( \alpha \))** = 0.10 (90% confidence level)
- **Hypothesized Population Mean (\( \mu \))** = 0.30 seconds

---

### **Step 3: Compute the Test Statistic (t-value)**

The formula for the t-statistic is:

\[
t = \frac{\bar{x} - \mu}{s / \sqrt{n}}
\]

Substitute the given values:

\[
t = \frac{0.25 - 0.30}{0.05 / \sqrt{30}} = \frac{-0.05}{0.05 / 5.477} = \frac{-0.05}{0.0091} \approx -5.49
\]

---

### **Step 4: Degrees of Freedom and Critical Value**

- **Degrees of Freedom (\( df \))**: For a one-sample t-test, \( df = n - 1 = 30 - 1 = 29 \).
- Using a **t-distribution table** or **t-distribution calculator**, the **critical t-value** for \( df = 29 \) and \( \alpha = 0.10 \) (two-tailed test) is approximately **±1.699**.

---

### **Step 5: Compare t-statistic with Critical Value**

- **Computed t-value** = **-5.49**
- **Critical t-value** = **±1.699**
