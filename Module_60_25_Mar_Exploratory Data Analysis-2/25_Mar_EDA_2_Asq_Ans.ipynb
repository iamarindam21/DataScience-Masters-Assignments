#Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (Assuming it's already uploaded as 'google_playstore.csv')
df = pd.read_csv('google_playstore.csv')

# Clean up or preprocess data if necessary (e.g., removing missing values)
df = df.dropna(subset=['Category', 'Rating'])  # Remove rows with missing values in 'Category' or 'Rating'

# Convert Rating to numeric (if needed) to avoid any issues
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Create a boxplot to compare ratings across app categories
plt.figure(figsize=(12, 8))
sns.boxplot(x='Category', y='Rating', data=df)
plt.xticks(rotation=90)  # Rotate the category names for better readability
plt.title('Distribution of Ratings by App Category')
plt.xlabel('App Category')
plt.ylabel('App Rating')

# Show the plot
plt.tight_layout()
plt.show()
```

# Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may impact your analysis.



```python
# Check for missing values in the dataset
missing_values = df.isnull().sum()

# Display columns with missing values
missing_values = missing_values[missing_values > 0]

# Show the missing values count for each column
print("Missing values per column:")
print(missing_values)

# Optionally, display the percentage of missing values
missing_percentage = (missing_values / len(df)) * 100
print("\nPercentage of missing values per column:")
print(missing_percentage)
```

### Impact of Missing Values:

1. **Loss of Information**: Missing values may lead to a loss of valuable data, especially if a large proportion of data points are missing for certain columns or rows. This could reduce the overall quality of the analysis.

2. **Bias in Analysis**: If the missing data is not missing at random, it can introduce bias into the analysis. For example, if higher-rated apps are more likely to have missing values, the analysis could incorrectly suggest that the overall app rating is lower than it should be.

3. **Skewed Results**: When missing values are not handled correctly, they can skew the results of any statistical analysis or machine learning model. This is especially problematic when analyzing correlations, regressions, or any models where complete data is crucial.

4. **Imputation or Removal**: If missing data is minimal, imputation techniques like filling missing values with the mean, median, or mode might be acceptable. However, if the missing data is extensive, it might be more appropriate to remove rows or columns containing the missing data to maintain the integrity of the dataset.

5. **Influence on Model Performance**: Machine learning models require complete data for training. Missing values may lead to reduced model accuracy, especially if no preprocessing steps are taken to handle them. Models like decision trees might be more robust to missing data, but others like linear regression may fail if the data is not properly preprocessed.

6. **Inconsistent Patterns**: Missing data can indicate issues with data collection or inconsistencies in the dataset, which may affect the credibility of the findings. Proper handling of missing values is necessary to ensure the integrity of the analysis and any subsequent conclusions.

# Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset (assuming it's already in a pandas DataFrame called df)
# Clean the 'Price' column by removing the '$' sign and converting to numeric
df['Price'] = df['Price'].replace('₹', '', regex=True)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Group by 'Type' (Free or Paid) and calculate the mean of 'Price'
avg_price_by_type = df.groupby('Type')['Price'].mean().reset_index()

# Plotting the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='Type', y='Price', data=avg_price_by_type)
plt.title('Average Price by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price (in ₹)')
plt.show()
```
